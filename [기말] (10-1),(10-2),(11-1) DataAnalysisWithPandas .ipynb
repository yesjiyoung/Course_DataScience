### merge()

```python
# 두 개의 데이터셋이 based on the value of sinlge column일 경우 사용(?)
```

```python
df_A_B=pd.DataFrame({'month': ['Jan',"Feb", 'Marach', 'April'],
                  'Product_A' : [100,15.,200,130],
                  'Product_B': [90,110,140,170]})
```

```python
df_A_B
```

```python
df_C_D=pd.DataFrame({'month': ['Jan',"Feb", 'Marach', 'April'],
                  'Product_A' : [112,141,203,134],
                  'Product_B': [90,110,140,170]})
```

```python
# how : indicates the method of merging
# on indicates the key variable



# how : 3 개의 방식이있음
# -> 'left' : we merge the dataset based on the left data
# -> 'right'
# -> 'outer' : index가 합집합으로 다 ~~ 됨!!
# -> 'inner' : index가 교집합으로 축소됨!
df_A_B.merge(df_C_D, how='outer', on='month')
```

```python
df_left=pd.DataFrame({'key':['A','B','C'], 'left' : [1,2,3]})
```

```python
df_right=pd.DataFrame({'key':['B','A','D'], 'right' : [4,5,6]})
```

```python
df_left.merge(df_right, how='left', on='key')
```

```python
df_left.merge(df_right, how='right', on='key')
```

```python
df_left.merge(df_right, how='outer', on='key')
```

```python
df_left.merge(df_right, how='inner', on='key')
```
