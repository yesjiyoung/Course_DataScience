#### grid

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title of figure')
plt.grid(True)
```

#### Legend

```python
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
plt.plot(x,y1,'or-',x,y2,'>y:',x,y3,'sg--')
plt.legend(['data1','data2','data3'], loc='lower left') # loc = "best,center,upper left ,lower ..."
```

#### legend+label+grid

```python
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
plt.plot(x,y1,'or-',x,y2,'>y:',x,y3,'sg--')
plt.legend(['data1','data2','data3'], loc='lower left') # loc = "best,center,upper left ,lower ..."
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title of plot')
```

## Scatter plot

```python
# 두 개의 numerical data의 상관도 보기에 딱 좋은게 산전도임.
```

```python
height=[165,177,160,180,185]
weight=[62,67,55,74,90]
```

```python
plt.scatter(height, weight)
```

```python
plt.scatter(height, weight)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height & Weight')
plt.grid(True)
```

```python
## plt.scatter(x, y, s=size_n, c=colors, marker'marker_string', alpha=transparency) 알파값은 투명도!
```

```python
# Q. plot of city based on the size of population
# (x,y)
```

```python
city=['Seoul','Incheon','Gwangju','Taegu', 'Pusan']
lat=[37.56,37.45,35.16,35.87,35.18]
lon=[126.97, 126.70, 126.85, 128.6, 128.60]
pop=[16154,2751,2995,128.60,4454] #number/km^2
```

```python
pop_den=np.array(pop)*0.2
```

```python
color=['r','g','b','c','k']
plt.scatter(lon, lat, s=pop_den,c=color)
# plt.text(x,y, 'string')
# 텍스트 쓸려면 요로콤 쓰도록!!!!!!!!! 정확한 위치여 저게! 
plt.text(126.97, 37.56,'Seoul')
```

```python
color=['r','g','b','c','k']
plt.scatter(lon, lat, s=pop_den,c=color,alpha=0.3)
# plt.text(x,y, 'string')
for x,y,name in zip(lon, lat, city): #zip()함수
    plt.text(x,y,name)
plt.text(126.97, 37.56,'Seoul')
plt.xlabel('Logitude')
plt.ylabel('Latitude')
```

## bar chart

```python
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
```

```python
# ID before_diet after_diet
# M_1, 80, 70
# M_2, 90, 85
# M_3, 70, 67
```

```python
member_IDs=['m_01','m_02','m_03'] # ID
before_diet=[80,90,70]
after_diet=[70,85,67]
```

```python
# plt.bar(x,height) x : location number (numericl value) 
```

```python
import numpy as np
index=np.arange(3)
```

```python
plt.bar(index,before_diet)
```

```python
# 막대 넓이나 이런거 바꾸기
```

```python
#카테고리 value를 각 막대에 대입! 
```

```python
plt.bar(index,before_diet,tick_label=member_IDs,color='c',width=0.5)
#width in (0, 1)
```

```python
# 색 각자 바꾸기! 
colors=['b','r','g']
plt.bar(index,before_diet,tick_label=member_IDs,color=colors,width=0.5)
```

```python
# compare before diet with after diet
```

```python
# 인덱스 이름을 또 다르게 해줘야함@!!! 
plt.bar(index,before_diet,tick_label=member_IDs,color='r',width=0.3, align='edge',label='before'
       )
# index도 약간 거리를 줘야함. 
plt.bar(index+0.5,after_diet,tick_label=member_IDs,color='g',width=0.3, align='edge', label='after')
plt.legend()
plt.xlabel('ID')
plt.ylabel('Weight')
plt.title('Effect of diet')
```

### Histogram

- numerical data들을 시각화 하기에 좋음.

```python
math=[76,82,84,83,90,86,85,92,72,71,100,87,81,76,94,78,81,60,79,69,74]
```

## pie chart

- 카테고리 데이터들을 시각화 하기에 매우 좋음

```python
# plt.pie(x) 
```

```python
fruit=['apple','banana','strawberry','orange']
preference=[7,6,3,2]
```

```python
plt.pie(preference)
```

```python
# 이제 어떤 데이터가 어떤 카테고리 인덱스인지 알려줘야함
# label을 사용할 예쩡
```

```python
plt.pie(preference, labels=fruit, autopct='%0.2f')
# 0.1 = 1 decimal palce (소수점 첫째자리)
# 0.2 = 2 decimal palce (소수점 둘째자리)
# 0.2f%%
```
